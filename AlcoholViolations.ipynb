```python
import csv

with open("Alcoholic_Beverage_License_Violations.csv", encoding = 'latin-1') as file:
    alcohol = list(csv.reader(file))
```

```python
alcohol[:5]
```

```python
violations = [w[3] for w in alcohol[1:]]
```

```python
violations
```

```python
violation_counts = {}

for s in violations:
    if s not in violation_counts:
        violation_counts[s] = 1
    else:
        violation_counts[s] += 1
```
